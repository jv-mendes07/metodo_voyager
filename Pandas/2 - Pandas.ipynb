```python
# Importação do método nativo do Python .glob:
# Importação do pandas para facilitar na importação de múltiplos arquivos:
import glob
import pandas as pd
```

```python
# Atribuição do caminho 'content' para a variável path:
# Atribuição da englobação conjuntiva de todos os arquivos.csv que estão no caminho path:
# Print dos arquivos.csv que foram recolhidos:
path = r'/content/'
filenames = glob.glob(path + '/*.csv')
print(filenames)
```

```python
# Laço de repetição for para atribuir os dois arquivos.csv à uma lista que serão lidos
# como DataFrame pelo método do Pandas de leitura de csv:
lista_arquivos = []
for i in filenames:
  lista_arquivos.append(pd.read_csv(i))
```

```python
# Concatenção entre os dois arquivos.csv e atribuição à uma nova variável:
df_full = pd.concat(lista_arquivos, ignore_index = True)
```

```python
# Leitura das primeiras linhas de tal dataframe concatenado entre dois arquivos diferíveis:
df_full.head()
```

```python
# Leitura das últimas linhas de tal dataframe concatenado entre dois arquivos diferíveis:
df_full.tail()
```

```python
# Exemplo de criação de uma função para aplicarmos no dataframe com o método apply ou applymap:
def soma(a, b):
  return a + b
soma(2, 2)
```

```python
# Criação de uma função para realizarmos a subtração entre o valor máximo e o valor mínimo:
def sub_max_min(df):
  return df.max() - df.min()
```

```python
# Iremos separar do dataframe original a coluna 'Nome' para podermos ter somente colunas numéricas:
df2 = df.drop('Nome', axis = 1)
print(df2.shape)
df2
```

```python
# Aplicação da função de subtração entre o maior e o menor valor do dataframe sobre cada coluna:
df2.apply(sub_max_min, axis = 0)
```

```python
# Aplicação da função de subtração entre o maior e o menor valor do dataframe sobre cada linha:
df2.apply(sub_max_min, axis = 1)
```

```python
# Aplicação da função de subtração em cada coluna com uma função anônima lambda:
df2.apply(lambda x : x.max() - x.min(), axis = 0)
```

```python
# Aplicação da função de subtração em cada linha com uma função anônima lambda:
df2.apply(lambda x : x.max() - x.min(), axis = 1)
```

```python
# Função lambda aplicável para obtermos o retorno das primeiras letras de cada nome da coluna 'Nome':
df.Nome.apply(lambda x : x[0])
```

```python
# Função lambda aplicável para obtermos o retorno do quadrado de cada elemento numérico:
df2.apply(lambda x : x**2, axis = 0)
```

```python
# Importação do dataset do campeonato brasileiro:
df = pd.read_csv('/content/campeonato-brasileiro-full (1).csv')
print(df.shape)
df.head()
```

```python
# Amostra de 30 observações do dataset atribuída à uma variável:
df_sample  = df.sample(n = 30)
print(df_sample.shape)
df_sample.head()
```

```python
# Atribuição de colunas como índices para formar um dataframe multíndice:
df_sample.set_index(['Dia', 'Visitante Placar'], inplace = True)
df_sample.head()
```

```python
# Ordenação do índice do menor para o maior ou do maior para o menor:
# Ascending = True (ordem crescente) e False (ordem decrescente)
df_sample.sort_index()
```

```python
# Ordenação inversa dos dois índices, o Dia será ordenado do 'maior' (Segunda) para o 'menor' (Domingo) dia
# e Visitante Placar será ordenado do menor para o maior número de Visitante Placar:
df_sample.sort_index(ascending = [False, True], inplace = True)
df_sample.head()
```

```python
# Como resetar os índices, ou seja, retira-los do índice:
df_sample.reset_index(inplace = True)
df_sample.head()
```

```python
# Atribuição de colunas ao índice para termos novamente um dataframe multíndice:
df_sample.set_index(['Dia','Mandante Placar'], inplace = True)
df_sample.head()
```

```python
# Tal função irá inverter os índices, o primeiro será transformado em segundo índice e vice-versa:
df_sample.swaplevel()
```
