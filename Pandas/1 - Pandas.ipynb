```python
# Converter 'Data' de tipo object para tipo datetime:
df['Data'] = pd.to_datetime(df['Data'], format = '%Y-%m-%d')
df['Data'].dtype
```

```python
# Criação de uma coluna de ano em que o jogo foi realizado:
df['Ano_Jogo'] = df['Data'].dt.year
df.Ano_Jogo
```

```python
# Verificação em dataframe da coluna 'Ano_Jogo' adicionada:
df.head()
```

```python
# Atribuição à uma variável nova a filtragem somente de jogos ocorríveis no ano de 2002:
df_2002 = df[df['Ano_Jogo'] == 2002]
print(df_2002.shape)
df_2002.head()
```

```python
# Atribuição à uma nova variável os jogos ocorrentes entre os anos de 2000 até 2003 com o método between():
df_bet = df[df.Ano_Jogo.between(2000, 2003)]
print(df_bet.shape)
df_bet.tail()
```

```python
# Usando isin para atribuir à uma variável somente os jogos ocorrentes no intervalo dos anos de 2004 e 2005:
intervalo_ano = [2004, 2005]
df_04_05 = df[df.Ano_Jogo.isin(intervalo_ano)]
print(df_04_05.shape)
df_04_05.head()
```

```python
# Filtragem somente de jogos que não foram ocorríveis em 2004 ou em 2005 com o método .mask:
df_not_04_05 = df.mask((df['Ano_Jogo'] == 2004) | (df['Ano_Jogo'] == 2005))
print(df_not_04_05.shape)
df_not_04_05.head()
```
